```python
def f1(x):
  return x**2
def f2(x):
  return 2*x
def f3(x):
  return x+2

def comp(x0):
  x1 = f1(x0)
  print(f"x1={x1}")
  x2 = f2(x1)
  print(f"x2={x2}")
  x3 = f3(x2)
  print(f"x3={x3}")
  return x3

comp(2)
```

```python
def fdiff(f):
  def diff(x):
    h = 1e-6
    y = (f(x+h) - f(x-h))/(2*h)
    return y
  return diff

# f = lambda x: 2*x**2 + 2
f = lambda x: 1 / (1 + np.exp(-x))
f_dash = fdiff(f)

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-2, 2.1, 0.1)
y      = f(x)
y_dash = f_dash(x)
plt.plot(x, y,      label=r'y = f (x)', c='b')
plt.plot(x, y_dash, label=r"y = f'(x)", c='k')
plt.legend()
plt.show()
```
